## 變數影響範圍
+ 函數外的變數：
  + 函數內可以顯示該變數內容
  + 不屬於函數的區域內都可以使用
+ 函數內的變數：
  + 只在函數內產生效果，不會影響函數外的變數
  + 若函數內沒有進行變數宣告而進行改變內容動作將會產生錯誤訊息!

```python
a = 5
def fun_sum():
    a = 10
    print("函數內:",a)
print("函數外1:",a)
fun_sum()
print("函數外2:",a)
```

一個*代表以 tuple 的方式引入

兩個*代表以 dict 的方式引入

接收資料若需加入多個設定，這些設定請依照順序排列：

第一個參數設定為接收值

第三個參數設定為不固定值，可以是 tuple 或 dict 方式

```python
def printInfo(arg1,arg2="default",*arg3):
    print("arg1:",arg1)
    print("arg2:",arg2)
    print("arg3:",arg3)
    print("="*50)
printInfo(1)
printInfo(1,2)
printInfo(1,2,3)
printInfo(1,2,3,4)
printInfo(1,2,3,4,5)
```

## 匿名函數
不需要定義函數名稱，只需要用到運算式或表達分析語法。

Python使用lambda語法定義匿名函數。

匿名函數是一個表達式或計算式，並不是一個執行流程區塊。

匿名函數出現在一般函數不允許的地方，例如list裡面或著函數呼叫參數的位置。

```python
def f (x,y,z):
    return x + y + z
print(f(2,30,400))   

f2 = lambda x,y,z:x+y+z
print(f2(2,30,400))
```

```python
listFunc = [lambda x:x**2,lambda x:x**3,lambda x:x**4]
for f in listFunc:
   print(f(3))
```

```python
age = 17
msg = "成年" if age >= 18 else "未成年"
print(msg)
```

```python
findMin = lambda x,y:x if x < y else y
print(findMin(2,5))
print(findMin(6,3))
```

## Map

```python
list3 = [1,2,3,'',4,5,'']

def emptyToZero(data):
    result = []
    for v in data:
        if v == '':
            v = 0
        result.append(v) 
    return result
print(emptyToZero(list3))
        
```

```python
list4 = [1,2,3,4,5]
mapObj = map(lambda x : x + 1,list4)
print(list(mapObj))
```

```python
list3 = [1,2,3,'',4,5,'']

mapObj = map(lambda v:0 if  v == '' else v,list3)
print(list(mapObj))
```

```python
list4 = ["M","W","M","M","W"]
dirc = {"M":0,"W":1}
objMap = map(lambda x:dirc[x],list4)
print(list(objMap))
```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
import re
text = "Ken,vivin,lucy"
re1 = re.split(r",",text)
print(re1)
```

```python
import re
text = "one, two...ten"
re1 = re.split(r"[,. ]+",text)
print(re1)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456 "
rStr = re.sub(r"\d+","222",inputStr)
print(rStr)
rStr =re.sub(r"\d+","222",inputStr,count=1)
print(rStr)
```
