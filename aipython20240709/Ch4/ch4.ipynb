```python
txt = input("請輸入第一個字串:")
strList = []
strList.append(txt)
msg = txt
count = 0
inputMsg = "請輸入-"+txt[-1]+"-開始的字串"
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
        inputMsg = "請輸入-"+txt[-1]+"-開始的字串"
    else:
        inputMsg = "請再次輸入-"+txt[-1]+"-開始的字串"
        count+=1
    print("上一個字串是:"+msg)    
        
```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
print(list2)
del list2[2]
print(list2)
del list2[1:3]
print(list2)
```

```python
list1 = ['1', 'x', '2', 'x', '3']
del list1[:3]
print(list1)
```

## Dictionary

資料結構是由 key:value 所組成。

key 不能夠重複，否則會被後面的結果蓋過去。

可輸入 key 找尋您要找出來的值。

如果輸入的 key 不存在，那就會出現錯誤訊息。

key 只能是 字串 數字 tuple

```python
dict1 = {"a":100,"b":200,"c":300}
print(dict1)
print(dict1["b"])
print(dict1["a"])
print(dict1["x"])
```

```python
dict1 = {"a":100,"b":200,"b":300}
print(dict1)#key 重複，被後面的結果蓋過去
```

## key 與 value
+ key 不能於程式內改變：
  + 可以用數字、字串或者 tuple。
  + 不可以使用 list。
+ 如何找出所有的 key 與 value？
  + 您可以透過 dict1.keys( ) 這個方法找出所有的 key。
  + 您可以透過 dict1.values( ) 這個方法找出所有的 value。

```python
dict1 = {"a":100,"b":200,"c":300}
for key in dict1.keys():
    print(key,dict1[key])
print("="*50)    
for v in dict1.values():
    print(v)
print("="*50)  
for key in dict1:
    print(key,dict1[key])
```

## 新增與修改
+ 新增一筆資料
  + 請將 key 與 value 儲存至一個變數內。
  + 透過 dict.update(新增的資料) 這個方法的方式就可以新增。
+ 元素可否變更：
  + 以 = 指派方式指派給 key 就可以變更資料

```python
dict2 = {"ken":100,"vivin":95,"lucy":72}
print(dict2)
dict2.update({"vivin":82,"ken":99})#key存在修改
print(dict2)
dict2.update({"iris":63}) #key不存在變新增
print(dict2)
```

```python
dict2 = {"ken":100,"vivin":95,"lucy":72}
print(dict2)
dict2["vivin"] = 83#key存在修改
print(dict2)
dict2["iris"] = 77 #key不存在變新增
print(dict2)
```

## 自訂函數的運用

#### 程式設計時儘量減少相同語法的重複出現，為什麼?

修改時可能忘記其他地方沒有改，導致程式執行出現錯誤結果。

減少重複語法對於之後的維修會比較方便。


#### 要如何減少重複的相同語法?

把重複的相同語法當作工具看待，需要的時候呼叫他。

這種工具於程式語法內叫做函數(function)。


```python
for i in range(1,11):
    print(i,end=" ")
print()
for i in range(1,6):
    print(i,end=" ")
print()
for i in range(3,9):
    print(i,end=" ")
print()
```

```python
def func_sum():
    print("呼叫函式")
func_sum()    
```

```python
def func_sum(a,b):
    c = a + b
    print(c)
func_sum(2,3)    
```

```python
def func_loop(s,e):
    for i in range(s,e+1):
        print(i,end=" ")
    print()
func_loop(1,5)
func_loop(3,10)
func_loop(5,9)
```
