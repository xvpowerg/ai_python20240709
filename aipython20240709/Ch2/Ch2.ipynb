## 比較運算子

```python
a,b = 10,2
ans = a > b
print(a,">",b,"=",ans)
ans = a < b
print(a,"<",b,"=",ans)
ans = a >= b
print(a,">=",b,"=",ans)
ans = a <= b
print(a,"<=",b,"=",ans)
ans = a == b
print(a,"==",b,"=",ans)
ans = a != b
print(a,"!=",b,"=",ans)
```

## break 代表不回頭，跳出迴圈。
## continue 代表繼續執行下一個迴圈。

```python
for i in range(1,5):
    if i == 3:
        break
    print(i)
```

```python
for i in range(1,10):
    if i == 3 or i == 6:
        continue
    print(i,end=" ")
```

```python
##請輸入兩個數字
##第一筆數字 a1
##第二筆數字a2
#計算a1 * a2 輸出2 * 5 = 10
#請輸入是否繼續計算 輸入y就繼續計算 其他就結束計算




a1 = int(input("請輸入第一個數字"))
a2 = int(input("請輸入第二個數字"))
ans = a1 * a2
print(a1,"*",a2,"=",ans)

```

```python
n = "y"
while n == "y":
    a1 = int(input("請輸入第一個數字"))
    a2 = int(input("請輸入第二個數字"))
    ans = a1 * a2
    print(a1,"*",a2,"=",ans)
    n = input("是否繼續計算(y)")
```
