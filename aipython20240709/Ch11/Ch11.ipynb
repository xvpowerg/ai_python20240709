## 只是如果想要得到一條線性方程式，那該如何進行?

```python
import numpy as np
slope,intercept = np.polyfit(x,y,1)
print(slope,intercept)
```

```python
import numpy as np
slope,intercept = np.polyfit(x,y,1)
#y = ax + b
y1 = [slope * i + intercept for i in x ]
plt.plot(x,y,"--")
plt.plot(x,y1,"b")
plt.show()
```

```python
import numpy as np
slope1,slope2,intercept = np.polyfit(x,y,2)
print(slope1,slope2,intercept)
y1 = [slope1 * i**2 + slope2 * i + intercept    for i in x]
plt.plot(x,y,"--")
plt.plot(x,y1,"b")
plt.show()
```

```python
import pandas as pd
tvmarketing=pd.read_csv('tvmarketing.csv')
tvmarketing.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(tvmarketing["TV"],tvmarketing["Sales"])
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import sklearn.linear_model
print('我們先建立X與y資料，後續才能進行切割')
X=tvmarketing['TV']
y=tvmarketing['Sales']

print('準備做資料切割')
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)# 70% 做訓練 30%做測試
print(X_train.shape)
print(X_test.shape)
```

```python
model = sklearn.linear_model.LinearRegression()
import numpy as np
X_train = np.array(X_train).reshape(-1,1)#一維轉二維滿足 model.fit() 的要求
print(X_train.shape)
model.fit(X_train,y_train)#開始訓練
print(model.coef_)
print(model.intercept_)
```

```python
import matplotlib.pyplot as plt
#X2=np.array(X).reshape(1,-1)
X2=X#課後發現不需要轉換也可使用在此註解
y2=(model.coef_)*X2+model.intercept_ #線性方程
plt.scatter(X2,y)
plt.scatter(X2,y2,c='red')
plt.show()
```
