## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。

```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = "Python"
print(len(search1))

print(str1.find(search1))
index = str1.find(search1)
newStrat =  index+len(search1) + 1
print(str1[:index]+str1[newStrat:])
```

```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = input("請輸入搜尋字串:")
index = str1.find(search1)
if index == -1:
    print("找不到字串")
else:    
    print(str1.find(search1))
    index = str1.find(search1)
    newStrat =  index+len(search1)
    print(str1[:index]+str1[newStrat:])
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
print(str1.find(search1,4))#由index為4的字母開始找
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
index = 0
while True:
    index = str1.find(search1,index)
    if index == -1:
        break
    print("is:",index)
    index += len(search1)
```

# List VS Tuple

+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 可以隨時修改

## list 介紹
+ list 內資料用逗號隔開，放置於中括號的範圍內。
  + list_name = [1,2,3 ]
+ list 可以儲存不同類型的值或項目，如數字和字串。
  + list_name = [1, 2, ‘Python’, ‘lists’ ] 
+ list 內的資料可以進行增加、更新與移除。
+ 可透過.index( )方法取得指定資料的索引。

```python
list1 = ["H","i"]
tuple1 = ('H','i')
print(list1)
print(tuple1)
print(type(list1))
print(type(tuple1))

list1[1] = "X"
print(list1)

#tuple1[1] = "X"
```

```python
myList = ["Ken"]
print(myList)
myTuple = ("Lucy")
print(myTuple)
print(type(myTuple))
myTuple2 = ("Lucy",)
print(myTuple2)
print(type(myTuple2))
```

```python
#List與tuple合併
a = (3,1,2)
b = (6,7,8,9)
newTuple = a + b
print(newTuple)
list1 = [3,1,3]
list2 = ["A","B","C","D"]
newList = list1 + list2
print(newList)
```

```python
list1 = ["A","B","C","D","E","F"]
print(list1[2])
index = 3
print(list1[index])
print(list1[2:-1])
print(list1[::-1])
print(list1[-3:0:-1])
```

## 資料轉換
+ 轉為List 使用 list()函數
+ 轉為tuple 使用 tuple()函數

### 轉為List 

```python
r =range(1,10)
print(type(r))
list1 = list(r)
print(list1)
list2 = list("ABCDEF")
print(list2)
tuple1 = (1,2,3,4)
list3 = list(tuple1)
list3[2] = 10
print(list3)
```

### 轉為tuple

```python
r = range(1,10)
t1 = tuple(r)
print(t1)
list1 = ["A","B","C"]
t2 = tuple(list1)
print(t2)
```

```python
a = 10
b = 20
a,b = b,a
print(a,b)
x = b,a
print(type(x))
```

```python
data = ["Ken","Vivin","Joy"]
x,y,z = data#「解包」（unpacking）
print(x)
print(y)
print(z)
```

## 拷貝與附加

```python
list1 = [1,2,3,4,5]
list3 = list1
print(id(list1))
print(id(list3))
list3[1] = 99
print(list1)
```

```python
list1 = [1,2,3,4,5]
list4 = list1[:]
print(list4)
list4[2] = 99
print(list1)
print(id(list4))
print(id(list1))
```

```python
list1 = [1,2,3,4,5]
list5 = list1.copy()
print(list5)
print(id(list1))
print(id(list5))
```
